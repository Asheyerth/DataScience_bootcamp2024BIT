### Análisis superficial.
Se considera que las siguientes columnas son potencialmente interesantes para el manejo y análisis de datos del dataset escogido:  
- **location** La locación del usuario consultado. Permite realizar un análisis por locación.  
- **age** La edad del usuario consultado. Permite realizar un análisis por edad.  
- **rating** El puntaje dado por el usuario consultado al libro del registro.  
- **year_of_publication** Año en el que fue publicado el libro. Permite hacer un análisis por el año.  
- **publisher** Editorial que publicó el libro. Permite analizar si es relevante la reputación de la editorial.  
- **languaje** Lenguaje en el que se encuentra el libro. Permite analizar si es relevante el idioma de presentación.  
- **Category** Categoría en la que se encuentra el libro. Permite analizar si la categoría influye en el puntaje.  
 
  
Otras variables potencialmente relevantes son:  
 - **country - city - state** Son datos de locación del registro del usuario que dio el puntaje del libro. Estos tres datos permiten realizar análisis por sectores demográficos en relación con el puntaje.  
 - **user_id - isbn** Son los datos identitarios del usuario y el libro respectivamente.
